# 05 Business Model Development for AI-Based Products and Analysis of Ethical Challenges in Real-World AI Applications

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("HRDataset_v14.csv")

df.head()
```

```python
sensitive_data = df[['Sex', 'RaceDesc']].copy()
```

# Step 4: Remove Unnecessary & Sensitive Columns

```python
df_model = df.drop(columns=[
    'Employee_Name', 'EmpID', 'MaritalDesc', 'CitizenDesc',
    'HispanicLatino', 'RaceDesc', 'Sex'
], errors='ignore')
```

```python
df_model = pd.get_dummies(df_model, drop_first=True)
```

# Step 6: Feature Selection

```python
X = df_model.drop('Termd', axis=1)
y = df_model['Termd']
```

# Step 7: Train-Test Split (with Sensitive Data)

```python
X_train, X_test, y_train, y_test, sens_train, sens_test = train_test_split(
    X, y, sensitive_data, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
```

# 9: Train Model

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
results = sens_test.copy()

results['Actual'] = y_test.values
results['Predicted'] = y_pred
```

# Bias Summary

```python
bias_summary = results.groupby(['Sex', 'RaceDesc'])['Predicted'].mean().reset_index()

bias_summary.columns = ['Sex', 'Race', 'Avg Predicted Termination']

print(bias_summary)
```
