#  10 :Develop an AI model to analyze social media sentiment trends for financial markets.

# Step 1: Import Libraries

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, precision_score, recall_score
```

# Step 2: Load Dataset

```python
data = pd.read_csv("financial_sentiment.csv")
print(data.head())
```

```python
print(data.columns)
```

# Step 3: Convert Sentiment to Numbers

```python
data = data[data['Sentiment'] != 'neutral']

data['Sentiment'] = data['Sentiment'].map({'positive':1,'negative':0})
```

# Step 4: Define Input and Output

```python
X = data['Sentence']
y = data['Sentiment']
```

# Step 5: Split Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Step 6: Convert Text to Numbers

```python
vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

# Step 7: Train AI Model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# Step 8: Test Model

```python
y_pred = model.predict(X_test)
```

# Step 9: Evaluate Model

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```

# Step 10: Predict Sentiment

```python
text = ["Stock market will grow strongly"]

text_vector = vectorizer.transform(text)

prediction = model.predict(text_vector)
if prediction[0] == 1:
    print("Positive Market Sentiment")
else:
    print("Negative Market Sentiment")
```
