 # Step 1: Import Libraries

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

# Step 2: Load the Dataset

```python
data = pd.read_csv("Uber Customer Reviews.csv")
print(data.head())
```

# Step 3: Select Required Columns

```python
data = data[['content','score']]
print(data.head())
```

# Step 4: Remove Missing Values

```python
data = data.dropna()
```

# Step 5: Convert Rating into Sentiment

```python
data = data[data['score'] != 3]

data['sentiment'] = data['score'].apply(lambda x: 0 if x <= 2 else 1)

print(data.head())
```

 # Step 6: Define Input and Output

```python
X = data['content']
y = data['sentiment']
```

# Step 7: Split Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Step 8: Fit the Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

# Step 9: Train the AI Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

# Step 10: Test the Model

```python
y_pred = model.predict(X_test)
```

# Step 11: Evaluate the Model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```

# Step 12: Test With New Review

```python
review = ["Driver was very rude"]

review_vector = vectorizer.transform(review)

prediction = model.predict(review_vector)

if prediction[0] == 1:
    print("Positive Review")
else:
    print("Negative Review")
```
